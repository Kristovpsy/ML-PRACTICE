```python
from __future__ import absolute_import, division, print_function, unicode_literals


import tensorflow as tf

import pandas as pd
```

```python
CSV_COLUMN_NAMES = ['SepalLength', 'SepalWidth', 'PetalLength', 'PetalWidth', 'Species']
SPECIES = ['Setosa', 'Versicolor', 'Virginica']
# Lets define some constants to help us later on
```

```python
train_path = tf.keras.utils.get_file(
    "iris_training.csv", "https://storage.googleapis.com/download.tensorflow.org/data/iris_training.csv")
test_path = tf.keras.utils.get_file(
    "iris_test.csv", "https://storage.googleapis.com/download.tensorflow.org/data/iris_test.csv")

train = pd.read_csv(train_path, names=CSV_COLUMN_NAMES, header=0)
test = pd.read_csv(test_path, names=CSV_COLUMN_NAMES, header=0)
# Here we use keras (a module inside of TensorFlow) to grab our datasets and read them into a pandas dataframe
```

```python
train_y = train.pop('Species')
test_y = test.pop('Species')
train.head() # the species column is now gone
```

```python
def input_fn(features, labels, training=True, batch_size=256):
    # Convert the inputs to a Dataset.
    dataset = tf.data.Dataset.from_tensor_slices((dict(features), labels))

    # Shuffle and repeat if you are in training mode.
    if training:
        dataset = dataset.shuffle(1000).repeat()

    return dataset.batch(batch_size)
```

```python
# Feature columns describe how to use the input.
my_feature_columns = []
for key in train.keys():
    my_feature_columns.append(tf.feature_column.numeric_column(key=key))
print(my_feature_columns)
```

```python
# prompt: using keras instead of using estimator help me build the model

# Build the Keras model
model = tf.keras.models.Sequential([
  tf.keras.layers.Dense(128, activation='relu', input_shape=(len(train.keys()),)),
  tf.keras.layers.Dense(3, activation='softmax') # 3 output classes (Species)
])

# Compile the model
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

# Convert pandas dataframes to numpy arrays for Keras
X_train = train.values
y_train = train_y.values
X_test = test.values
y_test = test_y.values


# Train the model
model.fit(X_train, y_train, epochs=10)

# Evaluate the model
loss, accuracy = model.evaluate(X_test, y_test, verbose=0)
print('Test accuracy:', accuracy)
```

```python
import tensorflow_datasets as tfds
tfds.disable_progress_bar()

# split the data manually into 80% training, 10% testing, 10% validation
(raw_train, raw_validation, raw_test), metadata = tfds.load(
    'cats_vs_dogs',
    split=['train[:80%]', 'train[80%:90%]', 'train[90%:]'],
    with_info=True,
    as_supervised=True,
)
```

```python
get_label_name = metadata.features['label'].int2str  # creates a function object that we can use to get labels

# display 2 images from the dataset
for image, label in raw_train.take(2):
  plt.figure()
  plt.imshow(image)
  plt.title(get_label_name(label))
```

```python
IMG_SIZE = 160 # All images will be resized to 160x160

def format_example(image, label):
  """
  returns an image that is reshaped to IMG_SIZE
  """
  image = tf.cast(image, tf.float32)
  image = (image/127.5) - 1
  image = tf.image.resize(image, (IMG_SIZE, IMG_SIZE))
  return image, label
```

```python
train_batches = raw_train.map(format_example).batch(32).prefetch(1)
validation_batches = raw_validation.map(format_example).batch(32).prefetch(1) # define validation_batches here
test_batches = raw_test.map(format_example).batch(32).prefetch(1)
```

```python
IMG_SHAPE = (IMG_SIZE, IMG_SIZE, 3)

# Create the base model from the pre-trained model MobileNet V2
base_model = tf.keras.applications.MobileNetV2(input_shape=IMG_SHAPE,
                                               include_top=False,
                                               weights='imagenet')
```

```python
base_model.summary()
```

```python
base_model.trainable = False
```

```python
base_model.summary()
```

```python
global_average_layer = tf.keras.layers.GlobalAveragePooling2D()
```

```python
prediction_layer = keras.layers.Dense(1)
```

```python
model_catsdog = tf.keras.Sequential([
    base_model,
    global_average_layer,
    prediction_layer
])
```

```python
model_catsdog.summary()
```

```python
base_learning_rate = 0.0001
model_catsdog.compile(optimizer=tf.keras.optimizers.RMSprop(learning_rate=base_learning_rate), # Change 'lr' to 'learning_rate'
              loss=tf.keras.losses.BinaryCrossentropy(from_logits=True),
              metrics=['accuracy'])
```

```python
initial_epochs = 4
validation_steps=20

loss0,accuracy0 = model_catsdog.evaluate(validation_batches, steps = validation_steps)
```

```python
history = model_catsdog.fit(train_batches,
                    epochs=initial_epochs,
                    validation_data=validation_batches)

acc = history.history['accuracy']
print(acc)
```
